```python
import matplotlib.pyplot as plt
from matplotlib import style
style.use('ggplot')
import numpy as np
from sklearn.cluster import KMeans
```

```python
x=np.array([[1,2],
            [1.5,1.8],
            [5,8],
            [8,8],
            [9,11]])

plt.scatter(x[:,0], x[:,1], s=150)
plt.show()
```

```python
x= [1,2,3,4,5,6,7,8]
y= [2,4,6,8,10,12,14,16]

plt.scatter(x,y, s=100, c='r', marker='*')
plt.title('FOR FUN')
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.xlim(0,10)
plt.ylim(0,20)
plt.show()
```

```python
x= [1,2,3,4,5,6,7,8]
y= [2,4,6,8,10,12,14,16]

clf = KMeans(n_clusters=1)
clf.fit(x)
plt.scatter(x,y)
plt.show()
```
